```python
def contador (* num):
  tamanho = len (num)
  print (f"Rcebei os valores {num} que são ao todo {tamanho} numeros")


contador (2, 1, 7)
contador (8, 0)
contador (1, 2, 3, 4, 5, 6, 7, 8, 9)  
```

```python
#FUNCAO COM LISTAS
def dobra (lst):
  posicao = 0
  while posicao < len(lst):
    lst[posicao] *= 2
    posicao += 1

valores = [6, 3, 9, 1, 0, 2] #criando uma lista
dobra (valores) #passando uma lista como parâmetro
print (valores) #todas as funçoes em python são passadas como referência, entao muda o valor da variavel original   
```

```python
#97
#faça um programa que tenha uma função chamada escreva(), que receba um texto qualquer como parâmetro
#e mostre uma mensagem com o tamanho adaptável

def escreva (msgm):
  tamanho = len (msgm) + 4 #tamanho recebe o tamanho da palavra + 4 (pra ficar borda)
  print ("~" * tamanho)
  print (f"  {msgm}") #dois espaços no inicio para centralizar
  print ("~" * tamanho)

escreva ("Hello World")
```
