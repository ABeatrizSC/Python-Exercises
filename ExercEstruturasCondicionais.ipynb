```python
#10
print ("calcular imposto")
sal = float (input("Digite seu salário"))

imp = float
if sal <= 12.000:
  imp = sal*(10/100)
else: 
  imp = sal * (10/100) + (25/100) * (sal - 12000) 

print ("o imposto a ser pago será de", imp, "reais")    
```
