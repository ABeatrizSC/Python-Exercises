```python
#2
print ("algoritmo calcular area do terreno")
base = float (input("Digite o comprimento da base do terreno:"))
alt = float (input("Digite a altura do terreno:"))

area = base * alt

print ("A area do terreno é:", + area, "m2")
```

```python
#3
print ("Algoritmo converter celsius em Fahrenheit")
celsius = float (input("Digite a temperatura em graus celsius:"))

fahrenheit = (9/5) * celsius + 32

print ("A temperatura em Fahrenheit é:", + fahrenheit)
```
